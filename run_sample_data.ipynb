## Using sample data

Sample data from Illinois can be read in using `w4h.get_resources()`. By default, this will read in data from a medium-sized site in western Illinois. However, by setting the `scope='state'` parameter, this will instead read in data at a statewide level in Illinois. 

Also by default, `w4h.get_resources()` returns a dictionary of filepaths. This is faster. If instead you would like to return a dictionary with actual data objects, set `resource_type='data'`.

```python
import w4h
res=w4h.get_resources(resource_type='filepath', scope='local')
```

This is an example call of `wrh.run()` using the filepaths of the sample data, which were saved on your machine when you installed the w4h package.

Setting `verbose=True` allows you to follow what the program is doing and what parameters are being used.

```python
import w4h
res=w4h.get_resources(resource_type='filepath', scope='local')
results = w4h.run(well_data=res['well_data'],
                surf_elev_grid=res['surf_elev'],
                bedrock_elev_grid=res['bedrock_elev'],
                model_grid=res['model_grid'],
                study_area=res['study_area'],
                lith_dict = res['LithologyDict_Exact'],
                #lith_dict_start = res['LithologyDict_Start'],
                #lith_dict_wildcard = res['LithologyDict_Wildcard'],
                verbose=False
                )
```

```python
xSecPlot = w4h.plot_cross_section(results[1])
```
